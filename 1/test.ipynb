```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('asoria/bolivia-population')
df = dataset['train' ].to_pandas()


print(df.head())



# Data Cleanup
#df['job_posted_date'] =pd.to_datetime(df['job_posted_date'])
```

```python
# Verificar valores nulos
print(df.isnull().sum())

# Ejemplo de eliminación de filas con valores nulos
df = df.dropna()

# Ejemplo de relleno de valores nulos
df = df.fillna({'column_name': 'default_value'})
```

```python
df = df.drop_duplicates()
```

```python
# Ejemplo de corrección de valores inconsistentes en una columna
df['City'] = df['City'].str.strip().str.title()
```

```python
df.iloc[0:50] #acceder a una fila especifica,cuando pones : es decir de cual a cual quieres
```

```python
promedio = df['Population'].mean() #promedio de la columna
print(promedio)
```
